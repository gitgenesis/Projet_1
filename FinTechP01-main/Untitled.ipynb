```python
import datetime, re

datetime_pat = re.compile(r'datetime\.datetime\([\d, ]+\)')

def convert_datetime_values(d):
    for key, value in d.items():
        if datetime.pat.match(value):
            d[key] = eval(value)
    return d
```
